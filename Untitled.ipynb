```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read csv file
df=pd.read_csv("Diwali Sales Data.csv",encoding="unicode_escape")
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull()
```

```python
# Check null values
df.isnull().sum()
```

```python
# Drop unwanted column
df.drop(["Status","unnamed1"],axis=1,inplace=True)
```

```python

df.info()
```

```python
# Drop null values
df.dropna(inplace=True)
```

```python
df.info()
```

```python
# Change datatype
df["Amount"]=df["Amount"].astype(int)
```

```python
df.columns
```

```python
# rename Column
df.rename(columns={'Marital_Status':'Shaadi'})
```

```python
Exploratory Data Analysis
```

```python
Gender/order count
```

```python
cont=sns.countplot(x="Gender",hue="Gender",data=df)
for bar in cont.containers:
    cont.bar_label(bar)
```

```python
Sales
```

```python

sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Gender',y= 'Amount',hue="Gender" ,data = sales_gen)
```

```python
From above graph we can see that most of the buyers are female and even the purchasing power of females are greater than men
```

```python
Age/order count
```

```python
cont_age=sns.countplot(x="Age Group",hue="Gender",data=df)
for bar in cont_age.containers:
    cont_age.bar_label(bar)
```

```python
Age/ Sale
```

```python
sal_age=df.groupby(["Age Group"],as_index=False)["Amount"].sum().sort_values(by="Amount", ascending=False)
sns.barplot(x="Age Group",y="Amount", data=sal_age)
```

```python
From above graph we can see that most of the buyers are of age group between 26-35 yrs female
```

```python
State/total orders
```

```python
Sal_state=df.groupby(["State"],as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=False).head(10)
sns.set(rc={"figure.figsize":(15,5)})
cont=sns.barplot(x="State",y="Orders",data=Sal_state)
for bar in cont.containers:
    cont.bar_label(bar)
```

```python
Sale From top 10 States
```

```python
Sal_state=df.groupby(["State"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(10)
sns.set(rc={"figure.figsize":(15,5)})
cont=sns.barplot(x="State",y="Amount",data=Sal_state)
for bar in cont.containers:
    cont.bar_label(bar)
```

```python
From above graphs we can see that most of the order & total sales are from uttar Pradesh, Maharashtra and Karnataka respectively.
```

```python
Marital Status/Count
```

```python
df["Marital_Status"].replace([0,1],["Married","Unmarried"])
sns.set(rc={"figure.figsize":(7,5)})
cont=sns.countplot(x="Marital_Status",data=df)
for bar in cont.containers:
    cont.bar_label(bar)
```

```python
Sal_Marital=df.groupby(["Marital_Status","Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(10)
sns.set(rc={"figure.figsize":(7,5)})
cont=sns.barplot(x="Marital_Status",y="Amount",hue="Gender",data=Sal_Marital)
```

```python
From above graphs we can see that most of the buyers are married(women) and they have high purchasing power.
```

```python
Product Category/Order count
```

```python
sns.set(rc={"figure.figsize":(22,5)})
cont=sns.countplot(x="Product_Category",data=df)
for bar in cont.containers:
    cont.bar_label(bar)
```

```python
Sal_Category=df.groupby(["Product_Category"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(10)
sns.set(rc={"figure.figsize":(20,5)})
cont=sns.barplot(x="Product_Category",y="Amount",data=Sal_Category)
```
